## Instacart

```python
#Import products Dataset
i_products_data = pd.read_csv("Data/instacart/products.csv", encoding = 'utf8')
#print(i_products_data.apply(lambda x: pd.lib.infer_dtype(x.values)))

#Import Prior Products (Transactions 1) Dataset
i_trans_t_data = pd.read_csv("Data/instacart/order_products__train.csv", encoding = 'utf8')
#print(i_trans_data.apply(lambda x: pd.lib.infer_dtype(x.values)))

#Import Training Products (Transacitons 2) Dataset
i_trans_p_data = pd.read_csv("Data/instacart/order_products__prior.csv", encoding = 'utf8')
#print(i_trans_data.apply(lambda x: pd.lib.infer_dtype(x.values)))

#Import Order Dataset
i_order_data = pd.read_csv("Data/instacart/orders.csv", encoding = 'utf8')
#print(i_order_data.apply(lambda x: pd.lib.infer_dtype(x.values)))

#Import Aisle Names
i_aisle_data = pd.read_csv("Data/instacart/aisles.csv", encoding = 'utf8')
#print(i_order_data.apply(lambda x: pd.lib.infer_dtype(x.values)))
```

```python
#Merge Prior and Training Products Dataset

i_trans_data = pd.concat([i_trans_t_data, i_trans_p_data])
```

```python
i_order_data.head()
```

```python
i_trans_data.head()
```

```python
#Add trans data to order data

trans_order_data = pd.merge(i_order_data, i_trans_data, on="order_id")
trans_order_data
```

```python
#Take a random sample of 100 users:
user_values = i_order_data.user_id.values
clean_user_values = np.unique(user_values)
clean_user_values
```

```python
#Take random sample of 100 users
np.random.seed(161803399)
rand_user_values = np.random.choice(clean_user_values, 100)

rand_user_values_sort = np.sort(rand_user_values)
print(rand_user_values_sort)
```

```python
#Create a new dataframe for those 100

i_to_100 = trans_order_data.loc[trans_order_data["user_id"].isin(rand_user_values_sort)]
i_to_sorted = i_to_100.sort_values("user_id")
i_to_sorted
```

```python
i_to_sorted.shape
```

```python
#Create New Dataset by merging Transaction/Order and Product by index of Product_ID

top_i_set = pd.merge(i_to_sorted, i_products_data, on="product_id")
top_i_set
```

```python
#Now merge with aisle data:

final_i_set = pd.merge(top_i_set, i_aisle_data, on="aisle_id")
```
